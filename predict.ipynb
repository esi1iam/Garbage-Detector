```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = tf.keras.models.load_model('models/imageclassifier.h5')
```

```python
from tensorflow.keras.preprocessing import image
import cv2
def waste_prediction(IMG):
  img = cv2.imread(IMG)
  resize = tf.image.resize(img, (256,256))
  plt.imshow(resize.numpy().astype(int))
  plt.show()
  yhat = model.predict(np.expand_dims(resize/255, 0))
  print(yhat)
  if yhat > 0.5: 
    print(f'The Place Is Clean')
  else:
    print(f'The PLace Contains Garbage')
```

```python
waste_prediction("test/w1.jpg")
waste_prediction("test/w2.jpg")
waste_prediction("test/w3.jpg")
waste_prediction("test/w4.jpg")
waste_prediction("test/w5.jpg")
waste_prediction("test/w6.jpg")
waste_prediction("test/w7.jpg")
waste_prediction("test/w8.jpg")
waste_prediction("test/w9.jpg")
waste_prediction("test/w10.jpg")
```
